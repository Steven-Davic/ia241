# What are the average cases per month

It is easy for analyst (and myself) to get lost in the data. That's why I decided to ask what the average cases per month was so that I could have fewer data points and see an overall trend line emerge.

```python
avg_case_per_month = df.groupby('month').mean()['cases']
```

```python
avg_case_per_month
```

```python
avg_case_per_month.plot()
```

It is notable to view the curve and its slope. Keep in mind, the data after "Month 4" is not avalible yet, so the declining slope is not visualizing anything but a steady return to 0 cases.
